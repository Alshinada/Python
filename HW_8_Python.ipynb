```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np
```

```python
#Задача №1:
x = np.linspace(0,5,50)
print(x)
print()

y = np.exp(x)
print(y)
print()

fig = plt.figure()
y = np.exp(x)
plt.plot(x, y)
```

```python
#Задача №2:
x = np.linspace(1,10)
print(x)
```

```python
y=np.log(x)
print(y)
```

```python
y1 = np.log2(x)
print(y1)
```

```python
y2 = np.log10(x)
print(y2)
```

```python
fig = plt.figure()
plt.plot(x, y, 'r')
plt.plot(x, y1, '--y')
plt.plot(x, y2, ':g')
```

```python
#Задача №3:
x = np.linspace(0,1)
x
```

```python
y = np.arcsin(x)
print(y)
```

```python
y1 = np.arccos(x)
print(y1)
```

```python
#Задача №4:
x = np.linspace(0,10)
x
```

```python
y = np.sinh(x)
x
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y)
ax.set(xlim=(-2,12), ylim=(-1000, 12000), xlabel='Линейный массив', ylabel='sinh(x)', title='График гиперболического синуса')
```

```python
#Задача №5:
fig, ax = plt.subplots(2,3)

ax[0][0].plot(np.linspace(1,10),np.log(x))
ax[0][1].plot(np.linspace(1,10),np.log2(x))
ax[0][2].plot(np.linspace(1,10),np.log10(x))

ax[1][0].plot(np.linspace(0,1),np.arcsin(x))
ax[1][1].plot(np.linspace(0,1),np.arccos(x))
ax[1][2].plot(np.linspace(0,1),np.arctan(x))
```
