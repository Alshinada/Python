```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('classic')

import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 500)
y = np.sin(x)
y1 = np.cos(x)

plt.plot(x, y)
plt.plot(x, y1)
```

```python
import seaborn as sns
sns.set()

plt.plot(x, y)
plt.plot(x, y1)
```

```python
sns.set_style("white")

plt.plot(x, y)
plt.plot(x, y1)
```

```python
sns.set_style("dark")

plt.plot(x, y)
plt.plot(x, y1)
```

```python
sns.set_style("whitegrid")

plt.plot(x, y)
plt.plot(x, y1)
```

```python
sns.set_style("darkgrid")

plt.plot(x, y)
plt.plot(x, y1)
```

```python
sns.set_style("ticks")

plt.plot(x, y)
plt.plot(x, y1)
```

```python
sns.set_style("whitegrid", {'grid.color':'green'})

plt.plot(x, y)
plt.plot(x, y1)
```

```python
sns.set_style("whitegrid")

z=1

a = ['z', 'sin(z)', 'cos(z)', 'tan(z)']
b = [z, np.sin(z), np.cos(z), np.tan(z)]

sns.barplot(x=a, y=b)
```

```python
sns.set_style("whitegrid")

z=1

a = ['z', 'sin(z)', 'cos(z)', 'tan(z)']
b = [z, np.sin(z), np.cos(z), np.tan(z)]

sns.barplot(x=b, y=a)
```

```python
sns.set_style("whitegrid")

z=1

a = ['z', 'sin(z)', 'cos(z)', 'tan(z)']
b = [z, np.sin(z), np.cos(z), np.tan(z)]

sns.barplot(x=b, y=a, color='red')
```
