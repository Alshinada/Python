```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
#pervy sposob
fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x), '--')
```

```python
#vtoroy sposob
fig = plt.figure()
y=np.sin(x)
y1=np.cos(x)
plt.plot(x, y)
plt.plot(x, y1, '--')
```

```python
fig.savefig('sin_cos.png')
```

```python
fig, ax = plt.subplots(2)

ax[0].plot(x,y)
ax[1].plot(x,y1)
```
